```python
import h5py
import matplotlib.pyplot as plt
import numpy as np

cellNumber = 100
simLength = 15 #in seconds
freq = 8 #in Hz
max_jitter = 20 #in ms

delay = 1/freq

def new_jitter(arr,max_jitter):
    for i in range(len(arr)):
        jitter = np.random.randint(low=(max_jitter*-1), high=max_jitter,size=1,dtype=int)
        jitter = jitter/1000
        arr[i] = arr[i] + jitter
    return arr

total_timestamps = []
total_node_ids = []
for i in range(cellNumber):
    timestamp = np.arange(0,simLength,delay)
    timestamp = new_jitter(arr=timestamp,max_jitter=max_jitter)
    total_timestamps.extend(timestamp)
    for j in range(len(timestamp)):
        total_node_ids.append(i)

total_timestamps= [x * 1000 for x in total_timestamps] # Convert from S to ms
plt.scatter(total_timestamps,total_node_ids)
plt.xlim(0,2000)
plt.show()
```
